```python
import datetime
import hashlib
class block:
    def __init__(self,index,time,data,prehash):
        self.index=index
        self.time=datetime.datetime.now()
        self.data=data
        self.prehash=prehash
    def get_block_hash(self):
        hash_code = "{}{}{}{}".format(self.index, self.time, self.data, self.prehash)
        x= hashlib.sha256(hash_code.encode('utf8')).hexdigest()
        return x
    
def genesis_block():
    b=block(0,datetime.datetime.now(),'hello',0)
    h=b.get_block_hash()
    a=[b.index,b.time,b.data,b.prehash,h]
    return a
    

def create_block(index,time,data,prehash):
    b=block(index,time,data,prehash)
    ph=b.get_block_hash()
    a=[b.index,b.time,b.data,b.prehash,ph]
    return a

def proof_of_work(c):
        
    nonce = 0
    difficulty=2
    while not c.startswith('0' * difficulty):
        nonce += 1
        hash_code="{}{}".format(nonce,c)
        c=hashlib.sha256(hash_code.encode('utf8')).hexdigest()
    return c

    
chain=[]
print ('enter no of blocks :')
n=int(input())
ph=0
a=[]
for i in range(0,n):
  if i==0 :
        a=genesis_block()
        chain.append(a) 
        ph=a[4]
  else :
        print ('enter text\n')
        text=input ()
        a=create_block(i,datetime.datetime.now(),text,ph)
        print(a[3])
        c=a[3]
        proof_of_work(c)
        chain.append(a)
        ph=a[4]
print (chain)      

```
